```r

# For nicer printing
options(digits=2);

# Read in the data
data_mp3 <- read.csv("TechSurvey - Survey.csv",header=T); # 1353 obs of 82 vars



#convert date to unix second
for (i in c("Start", "End")) 
    data_mp3[,i] = as.numeric(as.POSIXct(strptime(data_mp3[,i], "%Y-%m-%d %H:%M:%S")))
for (i in 0:12){
    vnam = paste(c("PG",i,"Submit"), collapse="")
    data_mp3[,vnam] = as.numeric(as.POSIXct(strptime(data_mp3[,vnam], "%Y-%m-%d %H:%M:%S")))
}
#calculate differences in time    
for (i in 12:0){
    pv = paste(c("PG",i-1,"Submit"), collapse="");
    if (i==0) 
        pv="Start";
    vnam = paste(c("PG",i,"Submit"), collapse="");
    data_mp3[,vnam] = data_mp3[,vnam] -data_mp3[,pv];
}


```

```r
#Summarize all data
summary(data_mp3);
```
